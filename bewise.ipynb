```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
q = """SELECT dlg_id, line_n,df.role, text, 
    MAX(
       case  when (role='manager' and (text like '%здравствуй%' or text like '%добрый%' or text like '%привет%'))
             then 1 else 0 end
        ) OVER(PARTITION BY dlg_id) as cnt_greet,
    MAX(
        case  when (role='manager' and (text like '%здравствуй%' or text like '%добрый%' or text like '%привет%'))
             then text else '-' end
        ) OVER(PARTITION BY dlg_id) as txt_greet,
        
    MAX(
       case  when (role='manager' and (text like '%до свидания%' or text like '%до встречи%' 
       or text like '%доброго%' or text like '%на связи%' 
       or text like '%лучшего%' or text like '%хорошего%' or text like '%хороших%'))
             then 1 else 0 end
        ) OVER(PARTITION BY dlg_id) as cnt_farewell,
    MAX(
       case  when (role='manager' and (text like '%до свидания%' or text like '%до встречи%' or text like '%пока%'
       or text like '%доброго%' or text like '%на связи%' 
       or text like '%лучшего%' or text like '%хорошего%' or text like '%хороших%'))
             then text else '-' end
        ) OVER(PARTITION BY dlg_id) as txt_farewell,
        
    MAX(
       case  when (role='manager' and (text like '%меня зовут%' or text like '%меня%зовут%' or text like '%мое имя%'
       or text like '%я%компани%' or text like '%это%компани%' or text like '%вас%беспокоит%'))
             then 1 else 0 end
        ) OVER(PARTITION BY dlg_id) as cnt_name_manag,
    MAX(
       case  when (role='manager' and (text like '%меня%зовут%' or text like '%мое имя%'
       or text like '%я%компани%' or text like '%это%компани%' or text like '%вас%беспокоит%'))
             then text else '-' end
        ) OVER(PARTITION BY dlg_id) as txt_name_manag,
        
    MAX(
       case  when (role='manager' and (text like '% это % из %' or text like '% я % из %' or text like '%компани%'))
             then 1 else 0 end
        ) OVER(PARTITION BY dlg_id) as cnt_name_compan,
    MAX(
       case  when (role='manager' and (text like '% это % из %' or text like '% я % из %' or text like '% компани%'))
             then text else '-' end
        ) OVER(PARTITION BY dlg_id) as txt_name_compan,
        
   substr(
       max(
       case  when (role='manager' and (text like '% это % из %' or text like '% я % из %' or text like '% компани%'))
             then text else '-' end
           ) OVER(PARTITION BY dlg_id)   
       , 
       
       instr(       max(
       case  when (role='manager' and (text like '% это % из %' or text like '% я % из %' or text like '% компани%'))
             then text else '-' end
           ) OVER(PARTITION BY dlg_id) 
       ,'компани')+9, 15) AS name_compan
                                                    
                            
       FROM df
       order by df.dlg_id, df.line_n;"""

names = pysqldf(q)
names
```
